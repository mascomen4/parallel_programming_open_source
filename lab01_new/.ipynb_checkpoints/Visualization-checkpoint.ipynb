```python
# Before the import I've added by hand the row "mes" in the text. 
data = pd.read_csv("out.txt")
```

```python
data[:5]
```

```python
mes = data.to_numpy()
```

```python
plt.rcParams.update({
    "font.serif": ['Times New Roman'],
    "font.sans-serif": ['Arial'],
    "font.family": 'serif',
    "font.size": "4",
    # не использовать настройки шрифтов самого matplotlib
    "pgf.rcfonts": False,
    "pgf.texsystem": 'xelatex',
    "xtick.minor.visible": True,
    "ytick.minor.visible": True,
    "axes.grid": True,
    "axes.grid.which": 'both',
    "grid.linewidth": 0.25,
    "grid.linestyle": 'dashed',
    "grid.color": 'gray',
})
```

```python
std = np.std(mes)
avg = np.average(mes)
```

```python
# Гистограмма разниц по времени. 
%matplotlib inline


fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(1,1,1)
ax.hist(mes, density=True, bins=400, range=(0, 0.04), label=f"Гистограмма разниц по времени, \n среднее: {avg}, \
\nдисперсия: {std}")
ax.set_ylabel('Frequency')
ax.set_xlabel('Data')
ax.legend(loc="best")
```

```python
print("сохраняем изображения")
# Если каталога нет, то создаем
if os.path.exists(FOLDER):
    print(f"{FOLDER} существует")
else:
    print(f"{FOLDER} не существует, создаем")
    os.makedirs(FOLDER)

params = dict(format=FMT, dpi=300, bbox_inches='tight', pad_inches=0.0)
ax.savefig(f"{FOLDER}diff.{FMT}", **params)
```
